```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset
df = pd.read_csv('diabetes.csv')
```

# UNIVARIATE ANALYSIS


```python
# Outcome distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Outcome", data=df, palette="Set2")
plt.title("Outcome Distribution (0 = Non-diabetic, 1 = Diabetic)")
plt.show()
```

```python
# Histogram for numerical columns
num_cols = df.columns[:-1]   # exclude Outcome
df[num_cols].hist(figsize=(12,10), bins=20, color="skyblue", edgecolor="black")
plt.suptitle("Histograms of Features", y=1.02)
plt.show()
```

```python
# Boxplots (check outliers)
plt.figure(figsize=(15,10))
for i, col in enumerate(num_cols, 1):
    plt.subplot(3,3,i)
    sns.boxplot(y=df[col], color="lightgreen")
    plt.title(f"Boxplot - {col}")
```

```python
plt.tight_layout()
plt.show()
```

# BIVARIATE ANALYSIS


```python
# Outcome vs numerical features (boxplots)
plt.figure(figsize=(15,10))
for i, col in enumerate(num_cols, 1):
    plt.subplot(3,3,i)
    sns.boxplot(x="Outcome", y=col, data=df, palette="Set1")
    plt.title(f"{col} by Outcome")
plt.tight_layout()
plt.show()
```

```python
# Pairplot
sns.pairplot(df, hue="Outcome", diag_kind="kde", palette="husl")
plt.suptitle("Pairplot of Features by Outcome", y=1.02)
plt.show()
```

# CORRELATION ANALYSIS


```python
# Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
