```python
# Perform Exploratory Data Analysis for the dataset  Social Network Ads.csv using Matplotlib and Seaborn libraries.
```

```python
# import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
df = pd.read_csv("Social_Network_Ads.csv")
df.head()
```

```python
# Drop User ID (not useful for EDA)
if "User ID" in df.columns:
    df.drop("User ID", axis=1, inplace=True)
```

```python
df.columns
```

# UNIVARIATE ANALYSIS

```python
# Gender distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", data=df, palette="Set2")
plt.title("Gender Distribution")
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=20, kde=True, color="skyblue")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()
```

```python
# Salary distribution
plt.figure(figsize=(8,5))
sns.histplot(df["EstimatedSalary"], bins=20, kde=True, color="orange")
plt.title("Estimated Salary Distribution")
plt.xlabel("Estimated Salary")
plt.show()
```

```python
# Purchased distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Purchased", data=df, palette="Set1")
plt.title("Purchased Distribution (0 = No, 1 = Yes)")
plt.show()
```

# BIVARIATE ANALYSIS

```python
# Gender vs Purchased
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", hue="Purchased", data=df, palette="Set1")
plt.title("Purchase Decision by Gender")
plt.show()
```

```python
# Age vs Purchased
plt.figure(figsize=(8,5))
sns.boxplot(x="Purchased", y="Age", data=df, palette="Set2")
plt.title("Age Distribution by Purchase Decision")
plt.show()
```

```python
# Salary vs Purchased
plt.figure(figsize=(8,5))
sns.boxplot(x="Purchased", y="EstimatedSalary", data=df, palette="Set3")
plt.title("Salary Distribution by Purchase Decision")
plt.show()
```

```python
# Scatterplot Age vs Salary colored by Purchased
plt.figure(figsize=(8,5))
sns.scatterplot(x="Age", y="EstimatedSalary", hue="Purchased", data=df, palette="Set1")
plt.title("Age vs Salary (Purchased Highlighted)")
plt.show()
```

# MULTIVARIATE ANALYSIS


```python
# Pairplot of all features
sns.pairplot(df, hue="Purchased", palette="husl")
plt.suptitle("Pairplot of Features", y=1.02)
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
