```python
# Importing Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("Mall_Customers.csv")
```

# UNIVARIATE ANALYSIS


```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
print(df.columns)
```

```python
# Gender distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Genre", data=df, palette="Set2")
plt.title("Gender Distribution")
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=20, kde=True, color="skyblue")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()
```

```python
# Annual Income distribution
plt.figure(figsize=(8,5))
sns.histplot(df["Annual Income (k$)"], bins=20, kde=True, color="red")
plt.title("Annual Income Distribution")
plt.xlabel("Annual Income (k$)")
plt.show()
```

```python
# Spending Score distribution
plt.figure(figsize=(8,5))
sns.histplot(df["Spending Score (1-100)"], bins=20, kde=True, color="purple")
plt.title("Spending Score Distribution")
plt.xlabel("Spending Score (1-100)")
plt.show()
```

# BIVARIATE ANALYSIS

```python
# Age vs Spending Score
plt.figure(figsize=(8,5))
sns.scatterplot(x="Age", y="Spending Score (1-100)", hue="Genre", data=df, palette="Set1")
plt.title("Age vs Spending Score")
plt.show()
```

```python
# Income vs Spending Score
plt.figure(figsize=(8,5))
sns.scatterplot(x="Annual Income (k$)", y="Spending Score (1-100)", hue="Genre", data=df, palette="Set2")
plt.title("Annual Income vs Spending Score")
plt.show()
```

```python
# Boxplots for Gender vs Numerical Features
plt.figure(figsize=(15,5))
```

```python
plt.subplot(1,3,1)
sns.boxplot(x="Genre", y="Age", data=df, palette="Set3")
plt.title("Age by Gender")
```

```python
plt.subplot(1,3,2)
sns.boxplot(x="Genre", y="Annual Income (k$)", data=df, palette="Set3")
plt.title("Annual Income by Gender")
```

```python
plt.subplot(1,3,3)
sns.boxplot(x="Genre", y="Spending Score (1-100)", data=df, palette="Set3")
plt.title("Spending Score by Gender")
```

```python
plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

# MULTIVARIATE ANALYSIS

```python
# Pairplot of all numerical features
sns.pairplot(df.drop(columns=["CustomerID"]), hue="Genre", palette="husl")
plt.suptitle("Pairplot of Features", y=1.02)
plt.show()
```

```python
# Jointplot: Income vs Spending Score with regression line
sns.jointplot(x="Annual Income (k$)", y="Spending Score (1-100)", data=df, kind="reg", height=6, color="purple")
plt.show()
```

```python
# Age, Income, Spending Score together
plt.figure(figsize=(10,6))
sns.scatterplot(x="Annual Income (k$)", y="Spending Score (1-100)", size="Age", hue="Genre", data=df, sizes=(20,200), alpha=0.7, palette="Set1")
plt.title("Income vs Spending Score with Age as Size")
plt.show()
```
